```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#before binning
image = cv2.imread('clock_tower_image.jpg')
cv2_imshow( image)
/content/sample_data
```

```python
#after binning
binned_image = cv2.resize(image,(image.shape[1] // 2, image.shape[0] // 2), interpolation=cv2.INTER_AREA)
binned_image = cv2.resize(binned_image, (image.shape[1], image.shape[0]), interpolation=cv2.INTER_NEAREST)
cv2_imshow( binned_image)
```
